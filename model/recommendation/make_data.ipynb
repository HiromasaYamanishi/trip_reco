```python
from datetime import datetime
def correct_time(time):
    time = time[:time.index('月')+1]
    return time
def to_datetime(time):
    return datetime.strptime(time, '%Y年%m月')
df_jalan['visit_time_correct'] = df_jalan['visit_time'].apply(correct_time)
df_jalan['visit_time_correct'] = df_jalan['visit_time_correct'].apply(to_datetime)
```

```python
user = {}
all_user = []
all_spot = []
all_time = []
user_index = 0
for u,s,t in zip(df_jalan['url'], df_jalan['spot'], df_jalan['visit_time_correct']):
    if u not in user:
        user[u] = user_index
        user_index+=1
    if s in spot_id:
        ui = user[u]
        si = spot_id[s]
        all_user.append(ui)
        all_spot.append(si)
        all_time.append(str(t))
```

```python
from collections import defaultdict
def ten_core(df):
    thresh=5
    d_user = defaultdict(int)
    d_loc = defaultdict(int)
    for uid, lid in zip(df['userid'],df['spotid']):
        d_user[uid]+=1
        d_loc[lid]+=1

    d_user_10 = dict(filter(lambda item: item[1]>=10, d_user.items()))
    d_loc_10 = dict(filter(lambda item: item[1]>=5, d_loc.items()))
    df_ = df[df['userid'].isin(d_user_10.keys())]
    df_ = df_[df_['spotid'].isin(d_loc_10.keys())]
    return df_
```
